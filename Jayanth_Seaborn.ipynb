```python
#Importing Pckages 
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

### Group-1

```python
#Loading the Diabetes dataset.
Diabetes = pd.read_csv("C:\\Users\\Jay\\Interactive Data\\Diabetes.csv")
```

```python
# Create a scatter plot using seaborn's relplot
sn.relplot(x="Age", y="Insulin", data=Diabetes, kind="scatter")

# Show the plot
plt.show()
```

```python
# Create a scatter plot with "Age" as x-axis, "Insulin" as y-axis, and "Pregnancies" as hue and size
sn.relplot(x="Age", y="Insulin", hue="Pregnancies", size="Pregnancies", data=Diabetes, kind="scatter")

# Show the plot
plt.show()
```

```python
# Create a scatter plot with "Age" as x-axis, "Insulin" as y-axis, and "Pregnancies" as hue and size
sn.relplot(x="Age", y="Insulin", hue="Pregnancies", size="Pregnancies", data=Diabetes, kind="scatter")

# Set limits for the x and y axes to zoom in on the data
plt.xlim(20, 60)  # Adjust the values as needed for your data
plt.ylim(0, 300)   # Adjust the values as needed for your data

# Show the plot
plt.show()
```

### Group-2

```python
#Loading the salary datset.
salary = pd.read_csv("C:\\Users\\Jay\\Interactive Data\\Salary Data.csv")
# Create a scatter plot with "Age" on the x-axis and "Salary" on the y-axis
sn.scatterplot(x="Age", y="Salary", data=salary)
# Show the plot
plt.show()
```

```python
# Creating a scatter plot with "Age" on the x-axis and  "Salary" on the y-axis.Using "Education Level" for color (hue) and "Gender" for marker style
sn.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=salary)
# Set the title of the plot
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
# Show the plot
plt.show()
```

```python
# Setting  the Seaborn style style to white grid to remove grid lines
sn.set_style("whitegrid")

# Creating a scatter plot with "Age" on the x-axis and  "Salary" on the y-axis.Using "Education Level" for color (hue) and "Gender" for marker style
sn.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=salary)
# Set the title of the plot
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
# Customize the legend for better clarity
plt.legend(title='Legend', loc='upper left', bbox_to_anchor=(1, 1))
# Show the plot
plt.show()
```

```python
# Create a FacetGrid for separate scatter plots based on education level
g = sn.FacetGrid(salary, col="Education Level", hue="Gender")
g.map(sn.scatterplot, "Age", "Salary", marker="o", edgecolor="w")
g.add_legend()

# Show the plot
plt.show()
```

```python
#Other Vizualization apart from the scatterplot.
sn.barplot(x="Education Level", y="Salary", data=salary, ci=None)
plt.title('Bar Plot: Average Salary by Education Level')
plt.show()
```
