```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
cces_Examples = pd.read_csv("C:\\Users\\Jay\\Interactive Data\\cces_example_data.csv")
cces_Examples
```

```python
# figures with 2 axes
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# Set title for the first axis
ax1.set_title("Age Histogram")

# Set title for the second axis
ax2.set_title("Political Identity")

# Show the empty figure
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Create a bar plot for the 'Evaluation' variable
cces_Examples['Evaluation'].value_counts().sort_index().plot(kind='bar', ax=ax, color='skyblue')

# Set labels and title
ax.set_xlabel('Evaluation')
ax.set_ylabel('Count')
ax.set_title('Bar Plot of Evaluation Variable')

# Show the figure
plt.show()
```

```python

# Scatterplot

N = len(cces_Examples)  

x = cces_Examples['ConMediaUse'].values
y = cces_Examples['ImmiPhobia'].values

c = np.random.randint(1, 7, size=N) 

s = np.random.randint(10, 400, size=N)  

fig, ax = plt.subplots()

scatter = ax.scatter(x, y, c=c, s=s)

# produce color legend  
legend1 = ax.legend(*scatter.legend_elements(), 
                   loc="lower left",title="Classes") 
ax.add_artist(legend1)

# produce size legend
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)

legend2 = ax.legend(handles, labels, loc="upper right", title="Sizes") 

plt.xlabel('Conservative Media Use')
plt.ylabel('Immigration Phobia')
plt.title("Conservative Media Use vs Immigration Phobia")  

plt.show()

```
