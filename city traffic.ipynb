```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel("c:\\Users\\Jay\\Interactive Data\\Final Project\\Final_Majestic_to_AIM_jan-2016tomarch-2018.xlsx")
df
```

```python
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
#Created a 'Day_of_Week' column to differentiate between weekdays and weekends.
df['Day_of_Week'] = df['Date'].dt.day_name()
df
```

```python
time_of_day_categories = ['Early_Morning', 'AM', 'Midday', 'PM', 'Evening']
df['Time_of_Day'] = pd.cut(df['Date'].dt.hour, bins=[-1, 7, 12, 17, 20, 24], labels=time_of_day_categories, ordered=True)
```

```python
columns_to_keep = ['Date', 'Day_of_Week', 'Time_of_Day', 'Origin_Movement_ID', 'Origin_Display_Name', 'Destination_Movement_ID', 'Destination_Display_Name',
                  'Daily_Mean_Travel_Time_Seconds', 'AM_Mean_Travel_Time_Seconds', 'PM_Mean_Travel_Time_Seconds', 'Midday_Mean_Travel_Time_Seconds',
                  'Evening_Mean_Travel_Time_Seconds', 'Early_Morning_Mean_Travel_Time_Seconds']
df = df[columns_to_keep]
```

```python
df = df[['Date', 'Day_of_Week', 'Time_of_Day', 'Origin_Movement_ID', 'Origin_Display_Name', 'Destination_Movement_ID', 'Destination_Display_Name',
         'Daily_Mean_Travel_Time_Seconds', 'AM_Mean_Travel_Time_Seconds', 'PM_Mean_Travel_Time_Seconds', 'Midday_Mean_Travel_Time_Seconds',
         'Evening_Mean_Travel_Time_Seconds', 'Early_Morning_Mean_Travel_Time_Seconds']]
```

```python
df.head()
```

```python
# Plot the daily mean travel time
df.plot(x='Date', y='Daily_Mean_Travel_Time_Seconds', kind='line')
plt.title('Travel Time Variation Throughout the Day')
plt.xlabel('Date')
plt.ylabel('Travel Time (Seconds)')
plt.show()
```

```python
# Plot the travel time breakdown by time of day
fig, ax = plt.subplots(figsize=(12, 6))
df.plot(x='Date', y=['AM_Mean_Travel_Time_Seconds', 'PM_Mean_Travel_Time_Seconds', 'Midday_Mean_Travel_Time_Seconds', 'Evening_Mean_Travel_Time_Seconds', 'Early_Morning_Mean_Travel_Time_Seconds'], kind='line', ax=ax)
plt.title('Travel Time Variation by Time of Day')
plt.xlabel('Date')
plt.ylabel('Travel Time (Seconds)')
plt.legend(['AM', 'PM', 'Midday', 'Evening', 'Early Morning'])
plt.show()
```

```python
# Create a new column for day of the week
df['Day_of_Week'] = pd.to_datetime(df['Date']).dt.day_name()

# Plot the daily mean travel time, grouped by weekday vs. weekend
weekday_df = df[df['Day_of_Week'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])]
weekend_df = df[df['Day_of_Week'].isin(['Saturday', 'Sunday'])]

fig, ax = plt.subplots(figsize=(12, 6))
weekday_df.plot(x='Date', y='Daily_Mean_Travel_Time_Seconds', kind='line', ax=ax, label='Weekday')
weekend_df.plot(x='Date', y='Daily_Mean_Travel_Time_Seconds', kind='line', ax=ax, label='Weekend')
plt.title('Weekday vs. Weekend Travel Time Variations')
plt.xlabel('Date')
plt.ylabel('Travel Time (Seconds)')
plt.legend()
plt.show()
```
